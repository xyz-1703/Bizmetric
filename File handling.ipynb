```python
import os
import shutil


source_folder = input("Enter the folder path containing files: ")
small_folder = os.path.join(source_folder, "small")
medium_folder = os.path.join(source_folder, "medium")
large_folder = os.path.join(source_folder, "large")


os.makedirs(small_folder, exist_ok=True)
os.makedirs(medium_folder, exist_ok=True)
os.makedirs(large_folder, exist_ok=True)


for file_name in os.listdir(source_folder):
    file_path = os.path.join(source_folder, file_name)

    if os.path.isfile(file_path):
        size_in_kb = os.path.getsize(file_path) / 1024  

        if size_in_kb < 30:
            shutil.move(file_path, small_folder)

        elif 30 <= size_in_kb <= 100:
            shutil.move(file_path, medium_folder)

        else:
            shutil.move(file_path, large_folder)

print("Files have been organized successfully!")
```
